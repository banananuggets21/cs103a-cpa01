```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df
```

```python
df.shape
```

```python
#Global video game sales by platform per year
column = 'Global_Sales'

df2 = pd.pivot_table(df, values=column, index=["Year_of_Release"], columns=["Platform"])
df2
```

```python
df2.plot()
plt.title('Global Video Game Sales by Platform from 1980 to 2016')
```

```python
#Global Video Game Sales per year grouped by Publisher
gk = df.groupby('Publisher')
gk.first()
```

As identified in the above data, video game sales have increased over time, especially from publishers such as Nintendo, Take Two, EA and Ubisoft. Additionally, Playstation, WII and NES saw the greatest number of sales both during and even somewhat after its lifetime. This is most likely due to the increased storage space on newer consoles and the introduction of digital downloads on the PS4 and Xbox One (Nintendo Switch technically has it too, but it sucks and no one uses it), which allows players to download their games online and store it on their console's hard drive as opposed to on physical disks. There are exceptions to this, however, such as the Playstation VITA and WIIU, both of which are widely concidered to be failures in the gaming industry.

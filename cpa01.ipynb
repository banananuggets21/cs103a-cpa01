```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df
```

```python
df.shape
```

```python
df.plot()
plt.title('Global Video Game Sales from 1980 to 2016')
```

```python
column = 'Global_Sales'

df = pd.pivot_table(df, values=column, index=["Year_of_Release"], columns=["Platform"])
df
```

```python
gk = df.groupby('Platform')
gk.first()
```

As identified in the above data, video game sales have increased over time, especially from publishers such as Nintendo, Take Two, EA and Ubisoft. This is most likely due to the increased storage space on newer consoles and the introduction of digital downloads on the PS4 and Xbox One (Nintendo Switch technically has it too, but it sucks and no one uses it), which allows players to download their games online and store it on their console's hard drive as opposed to on physical disks. There are exceptions to this, however, such as the Playstation VITA and WIIU, both of which are widely concidered to be failures in the gaming industry.

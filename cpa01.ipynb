```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df = df.dropna(subset=['Name','Year_of_Release','Genre','Publisher'])
df
```

```python
#Prints number of rows and columns in the dataset
df.shape
```

```python
#Pivot Table of Global video game sales by platform per year
column = 'Global_Sales'

df2 = pd.pivot_table(df, values=column, index=["Year_of_Release"], columns=["Platform"])
df2
```

```python
#Plot of the big three consoles over the years
df2[['Wii','WiiU','PS', 'XOne', 'PS4', 'X360']].plot(figsize=(15,10))
plt.grid()
plt.title(column,fontsize=32)
```

```python
df2.plot()
plt.grid()
plt.title('Global Video Game Sales by Platform from 1980 to 2016')
```

```python
#Global Video Game Sales per year grouped by Publisher
gk = df.groupby('Publisher')
gk.first()
```

As identified in the above data, video game sales have increased over time, especially from publishers such as Nintendo, Take Two, EA and Ubisoft. It is fair to say this is likely due to improvements in graphics and processing power over time, which has improved accessibility to video games. Additionally, Playstation, WII and NES saw the greatest number of sales both during and even somewhat after its lifetime. It is also interesting to note there was a sudden decrease in sales towards the end of 2017. But since some 2020 data is also included, this could be due to the next gen consoles on backorder and the chip shortage due to the pandemic. This is most likely due to the increased storage space on newer consoles and the introduction of digital downloads on the PS4 and Xbox One, which allows players to download their games online and store it on their console's hard drive as opposed to on physical disks. There are exceptions to this, however, such as the Playstation VITA and WIIU, both of which are widely concidered to be failures in the gaming industry.
